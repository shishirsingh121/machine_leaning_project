```python
def changerange(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
    
    
    
```

```python
house.total_sqft=house.total_sqft.apply(changerange)
```

```python
house['price_per_sqft']=(house['price']*100000)/house['total_sqft']
```
