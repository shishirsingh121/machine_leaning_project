```python
X=clean_house_data.drop('price',axis=1)
```

```python
y=clean_house_data.price
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
column_tranformer=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
lr=LinearRegression()
scale=StandardScaler()
```

```python
pipe=make_pipeline(column_tranformer,scale,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,Y_pred)
```

```python
pipe=make_pipeline(column_tranformer,scale,Lasso())
pipe.fit(X_train,y_train)
Y_pred=pipe.predict(X_test)
r2_score(y_test,Y_pred)
```

```python
pipe=make_pipeline(column_tranformer,scale,Ridge())
pipe.fit(X_train,y_train)
Y_pred=pipe.predict(X_test)
r2_score(y_test,Y_pred)
```
